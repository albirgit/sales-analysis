## Objective
This notebook focuses on exploratory data analysis of the dataset
to understand structure, trends, and potential issues.

### Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Reading dataset

```python
df = pd.read_csv('../data/raw/Ecommerce_Sales_Data_2024_2025.csv')
df.head()
```

### Columns Datatypes

```python
df.info()
```

### Columns Mathematical Info

```python
df.describe()
```

### Null values count

```python
df.isna().sum()
```

### Highest Selling Category ( quantity wise )

```python
category_quantity = df.groupby('Category')['Quantity'].sum()
highest_qty_category = category_quantity.idxmax()
highest_qty = category_quantity.max()

print(f"Top Selling Category: {highest_qty_category}")
print(f"Total quantity: {highest_qty}")
```

### Insight: 
This shows that furniture has the highest selling quantity but more analysis required to show its revenue contribution as compared with other categories. 

### Highest Selling Category ( revenue wise )

```python
category = df.groupby('Category')['Sales'].sum()
highest_category = category.idxmax()
highest_category_qty = category.max()
print(f"highest Selling Category: {highest_category}")
print(f"Total sale: {highest_category_qty}")
```

### Insight:
This shows that home decor generates the highest revenue. Although it sells fewer units as compared to other categories but it has the highest revenue due to higher-priced products.

### Highest Selling Month

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Year-Month'] = df['Order Date'].dt.strftime('%Y-%m')

months_sales = df.groupby('Year-Month')['Sales'].sum()
highest_sales_month = months_sales.idxmax()
highest_sales_month_qty = months_sales.max()

print(f"highest Selling month: {highest_sales_month}")
print(f"Total quantity: {highest_sales_month_qty}")
```

### Insight:
This shows that May 2025 recorded the highest sales indicating a peak in costumer purchasing attitude due to some seasonal demand or some promotional campaigns.

### City with Highest Profit

```python
city_profit = df.groupby('City')['Profit'].sum()
highest_profit_city = city_profit.idxmax()
highest_profit_city_sales = city_profit.max()

print(f"City with highest profit: {highest_profit_city}")
print(f"Total quantity: {highest_profit_city_sales}")
```

### Insight: 
Bangalore generates the highest profit, indicating strong market performance and higher profitability in this city compared to other regions.

```python
category_sales = df.groupby('Category')['Sales'].sum()
plt.bar(category_sales.index, category_sales.values/10000000)
plt.title("Sales by Category")
plt.xlabel("Categories", fontsize = 13)
plt.ylabel("Sales (in Crores )", fontsize=13)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Insight:
Sales vary significantly across categories, with certain categories contributing a disproportionately higher share of total revenue. This indicates that business performance is driven more by high-value categories rather than uniform sales across all product types.

### Month vise Sales

```python
monthly_sales = df.groupby('Year-Month')['Sales'].sum()
plt.figure(figsize=(12,6))
plt.plot(monthly_sales.index, monthly_sales.values/10000000)
plt.title("Month vise Sales")
plt.xlabel("Months", fontsize = 13)
plt.ylabel("Sales in crores", fontsize = 13)
plt.xticks(rotation=90, ha = 'right')
plt.tight_layout()
plt.show()
```

### Insight:
Month-wise sales show noticeable fluctuations, indicating seasonal patterns in customer purchasing behavior. Certain months consistently record higher sales, which can be leveraged for targeted promotions and inventory planning.

### Payment mode usage Illustration

```python
payment_mode = df.groupby('Payment Mode').size()
plt.pie(payment_mode.values, labels = payment_mode.index, autopct = '%1.1f%%', startangle=90,
        colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'])
plt.title("payment mode usage illustration",fontsize = 15, fontweight = 'bold')
plt.show()
```

### Insight:
The payment mode distribution shows a clear preference for specific payment methods, indicating customer reliance on convenient and widely adopted payment options. This highlights the importance of optimizing and promoting popular payment channels.

### Region wise Sales

```python
region_wise_sales = df.groupby('Region')['Sales'].sum()
plt.barh(region_wise_sales.index, region_wise_sales.values/100000000)
plt.title("Region wise sales", fontsize = 13, fontweight = 'bold')
plt.xlabel("Sales in crores", fontsize = 13)
plt.ylabel("Region", fontsize = 13)
plt.show()
```

### Insight:
Sales performance differs across regions, with certain regions contributing a significantly larger share of total sales. This indicates regional variation in market demand and customer purchasing power.
